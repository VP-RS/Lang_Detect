```python
import pickle
import streamlit as st
```

```python
#1.Language Detection using transformer model
from transformers import pipeline

text = [
        "new jersey is sometimes quiet during autumn , and it is snowy in april",
        "votre moins aimé fruit est le raisin , mais mon moins aimé est la pomme"
       ]

model_ckpt = "papluca/xlm-roberta-base-language-detection"
pipe = pipeline("text-classification", model=model_ckpt)
pipe(text, top_k=1, truncation=True)
```

```python
# Save the model to a pickle file
with open('model.pkl', 'wb') as file:
    pickle.dump(pipe,file)
```

```python
# loading the trained model
pickle_in = open('model.pkl', 'rb')
classifier = pickle.load(pickle_in)
```

```python
#2.Language Translation using Transformer model(Pipeline API)
src = "en"
dst = "fr"

task_name = f"translation_{src}_to_{dst}"
model_name = f"Helsinki-NLP/opus-mt-{src}-{dst}"

translate  = pipeline(task_name, model=model_name, tokenizer=model_name)
```

```python
translate("california is usually quiet during march , and it is usually hot in june")[0]["translation_text"]
```

```python
# Save the model to a pickle file
with open('translate_model.pkl', 'wb') as file:
    pickle.dump(translate,file)
```

```python
# loading the trained model
pickle_in = open('translate_model.pkl', 'rb')
translator = pickle.load(pickle_in)
```
